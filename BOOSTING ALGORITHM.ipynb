```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
x=data.iloc[:,:-1]
```

```python
x
```

```python
y=data.iloc[:,-1]
y
```

```python
data_dmatrix=xgb.DMatrix(data=x,label=y)
```

```python
print(data_dmatrix)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
xg_reg=xgb.XGBRegressor(objective='reg:linear',comsample_bytree=0.3,learning_rate=0.1,max_depth=5,alpha=10,n_estimator=10)
#objective=to determine the loss function that is used to linear, it means regression.
#colsample=percentage of feature used per tree o.3=30%
#learning_rate=step size
#max_depth=determine tp how deeply each tree is allow to grow during boosting.
#alpha=regularization on weight
#n_estimators=no. of trees you want to build
```

```python
xg_reg.fit(x_train,y_train)
pred=xg_reg.predict(x_test)
```

```python
rmse=np.sqrt(mean_squared_error(y_test,pred))
print(rmse)
print(rmse/100)
```

```python
params={"objective":"reg:linear","comsample_bytree":0.3,"learning_rate":0.1,"max_depth":5,"alpha":10}
cv=xgb.cv(dtrain=data_dmatrix,params=params,nfold=3,num_boost_round=50,early_stopping_rounds=10,metrics="rmse",as_pandas=True,seed=123)
```
