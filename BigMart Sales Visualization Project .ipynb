```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dataset[['Item_Fat_Content','Item_Visibility','Item_MRP','Outlet_Establishment_Year']],dataset[['Item_Outlet_Sales']],test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
clf=RandomForestRegressor(n_estimators=100,max_depth=
                          50,min_samples_leaf=100)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(y_pred)
```

```python
accuracy=round(clf.score(x_train,y_train)*100)
accuracy
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor(max_depth=50,min_samples_leaf=100)
dtr.fit(x_train,y_train)
y_pred1=dtr.predict(x_test)
```

```python
print(y_pred1)
```

```python
accuracy1=round(dtr.score(x_train,y_train)*100)
accuracy1
```
