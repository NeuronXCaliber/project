```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
lb=LabelEncoder()
dataset.iloc[:,-1]=lb.fit_transform(dataset.iloc[:,-1].values)
ss=StandardScaler()
dataset.head()
```

```python
x=dataset.iloc[:,:21]
y=dataset.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train=ss.fit_transform(x_train)
x_test=ss.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import cross_val_score
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=10)
dtree.fit(x_train,y_train)
prediction1=dtree.predict(x_test)
prediction1
```

```python
count_missclassified1=(y_test!=prediction1).sum()
print(count_missclassified1)
accuracy1=metrics.accuracy_score(y_test,prediction1)
print(accuracy1)
cross_val_accur1=cross_val_score(DecisionTreeClassifier(criterion='entropy',max_depth=10),x_train,y_train)
print(cross_val_accur1)
```

```python
score1=cross_val_score(DecisionTreeClassifier(criterion='entropy',max_depth=10),x_train,y_train,cv=15)
print(score1)
```

```python
print(np.average(score1))
```

```python
cm1=metrics.confusion_matrix(y_test,prediction1)
print(cm1)
```

```python
c_r_dtree=metrics.classification_report(y_test,prediction1)
print(c_r_dtree)
```

```python
import graphviz
from sklearn import tree
dot_data=tree.export_graphviz(dtree,out_file=None,filled=True)
graph=graphviz.Source(dot_data,format='png')
graph
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(criterion='entropy',max_depth=10,n_estimators=50)
rfc.fit(x_train,y_train)
prediction2=rfc.predict(x_test)
prediction2
```

```python
count_missclassified2=(y_test!=prediction2).sum()
print(count_missclassified2)
accuracy2=metrics.accuracy_score(y_test,prediction2)
print(accuracy2)
cross_val_accur2=cross_val_score(RandomForestClassifier(criterion='entropy',max_depth=10,
                                                       n_estimators=50),x_train,y_train)
print(cross_val_accur2)
```

```python
score2=cross_val_score(RandomForestClassifier(criterion='entropy',max_depth=10,
                                                       n_estimators=50),x_train,y_train,cv=15)
print(score2)
```

```python
print(np.average(score2))
```

```python
c_m2=metrics.confusion_matrix(y_test,prediction2)
print(c_m2)
```

```python
c_r_rfc=metrics.classification_report(y_test,prediction2)
print(c_r_rfc)
```
