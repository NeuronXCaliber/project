```python
from sklearn.model_selection import train_test_split
x=dataset.iloc[:,:10]
y=dataset.iloc[:,-1]
print(x.head(2))
print(y.head(2))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import tree
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
#DecisionTree
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=10)
dtree.fit(x_train,y_train)
prediction1=dtree.predict(x_test)
print(prediction1)
```

```python
count_missclassified1=(y_test!=prediction1).sum()
print(count_missclassified1)
accuracy1=metrics.accuracy_score(y_test,prediction1)
print(accuracy1)
cross_val_accur1=cross_val_score(DecisionTreeClassifier(criterion='entropy',max_depth=10),
                                x_train,y_train)
print(cross_val_accur1)
```

```python
score1=cross_val_score(DecisionTreeClassifier(criterion='entropy',max_depth=10),x_train,y_train,cv=15)
print(score1)
```

```python
import numpy as np
print(np.average(score1))
```

```python
c_m1=metrics.confusion_matrix(y_test,prediction1)
print(c_m1)
```

```python
c_r1=metrics.classification_report(y_test,prediction1)
print(c_r1)
```

```python
#RandomForest
rfc=RandomForestClassifier(criterion='entropy',max_depth=10,n_estimators=50)
rfc.fit(x_train,y_train)
prediction2=rfc.predict(x_test)
print(prediction2)
```

```python
count_missclassified2=(y_test!=prediction2).sum()
print(count_missclassified2)
accuracy2=metrics.accuracy_score(y_test,prediction2)
print(accuracy2)
cross_val_accur2=cross_val_score(RandomForestClassifier(criterion='entropy',max_depth=10,n_estimators=50),
                                x_train,y_train)
print(cross_val_accur2)
```

```python
score2=cross_val_score(RandomForestClassifier(criterion='entropy',max_depth=10,n_estimators=50),
                                x_train,y_train,cv=15)
print(score2)
```

```python
print(np.average(score2))
```

```python
c_m2=metrics.confusion_matrix(y_test,prediction2)
print(c_m2)
```

```python
c_r2=metrics.classification_report(y_test,prediction2)
print(c_r2)
```

```python
#XGBClassification
xgb=XGBClassifier(max_depth=10,n_estimators=50,alpha=10,learning_rate=0.1)
xgb.fit(x_train,y_train)
prediction3=xgb.predict(x_test)
print(prediction3)
```

```python
count_missclassified3=(y_test!=prediction3).sum()
print(count_missclassified3)
accuracy3=metrics.accuracy_score(y_test,prediction3)
print(accuracy3)
cross_val_accur3=cross_val_score(XGBClassifier(max_depth=10,n_estimators=50,alpha=10,learning_rate=0.1),x_train,y_train)
print(cross_val_accur3)
```

```python
score3=cross_val_score(XGBClassifier(max_depth=10,n_estimators=50,alpha=10,learning_rate=0.1),x_train,y_train,cv=15)
print(score3)
```

```python
print(np.average(score3))
```

```python
c_m3=metrics.confusion_matrix(y_test,prediction3)
print(c_m3)
```

```python
c_r3=metrics.classification_report(y_test,prediction3)
print(c_r3)
```
