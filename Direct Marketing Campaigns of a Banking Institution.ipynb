```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import tree
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x=dataset.iloc[:,:16]
y=dataset.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train=ss.fit_transform(x_train)
x_test=ss.fit_transform(x_test)
```

```python
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=10)
dtree.fit(x_train,y_train)
prediction1=dtree.predict(x_test)
print(prediction1)
```

```python
count_missclassified1=(y_test!=prediction1).sum()
print(count_missclassified1)
accuracy1=metrics.accuracy_score(y_test,prediction1)
print(accuracy1)
cross_val_accur1=cross_val_score(DecisionTreeClassifier(criterion='entropy',max_depth=10),x_train,y_train)
print(cross_val_accur1)
```

```python
import numpy as np
score1=cross_val_score(DecisionTreeClassifier(criterion='entropy',max_depth=10),
                       x_train,y_train,cv=15)
print(score1)
print(np.average(score1))
```

```python
c_m1=metrics.confusion_matrix(y_test,prediction1)
print(c_m1)
```

```python
c_r_dtree=metrics.classification_report(y_test,prediction1)
print(c_r_dtree)
```

```python
rfc=RandomForestClassifier(criterion='entropy',max_depth=10,n_estimators=50)
rfc.fit(x_train,y_train)
prediction2=rfc.predict(x_test)
print(prediction2)
```

```python
count_missclassified2=(y_test!=prediction2).sum()
print(count_missclassified2)
accuracy2=metrics.accuracy_score(y_test,prediction2)
print(accuracy2)
cross_val_accur2=cross_val_score(RandomForestClassifier(criterion='entropy',max_depth=10,n_estimators=50),
                                x_train,y_train)
print(cross_val_accur2)
```

```python
score2=cross_val_score(RandomForestClassifier(criterion='entropy',max_depth=10,n_estimators=50),
                      x_train,y_train,cv=15)
print(score2)
```

```python
print(np.average(score2))
```

```python
import tensorflow as tf
from keras.layers import Flatten,Dense,Dropout,Activation
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.layers.advanced_activations import LeakyReLU
```

```python
model=Sequential()
model.add(Dense(256,input_dim=16,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.25))
model.add(Dropout(0.25))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(10))
model.add(LeakyReLU(0.1))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
es=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=3)
```

```python
history=model.fit(x_train,y_train,epochs=10,batch_size=256,validation_data=(x_test,y_test),callbacks=[es],
                 validation_split=0.25)
```

```python
print(model.evaluate(x_test,y_test))
```

```python
print(model.evaluate(x_train,y_train))
```

```python
predict1=model.predict(x_test,verbose=0)
print(predict1)
```

```python
predict2=model.predict_classes(x_test,verbose=0)
print(predict2)
```

```python
print(metrics.accuracy_score(y_test,predict2))
```

```python
print(metrics.precision_score(y_test,predict2,average='macro'))
```

```python
print(metrics.recall_score(y_test,predict2,average='macro'))
```

```python
print(metrics.confusion_matrix(y_test,predict2))
```

```python
print(metrics.classification_report(y_test,predict2))
```
